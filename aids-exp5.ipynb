```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense
```

```python
X = np.linspace(0, 50, 500)
y = np.sin(X)
```

```python
X_data = []
y_data = []
for i in range(10, len(y)):
    X_data.append(y[i-10:i])
    y_data.append(y[i])
X_data = np.array(X_data)
y_data = np.array(y_data)
X_data = np.reshape(X_data, (X_data.shape[0], X_data.shape[1], 1))
```

```python
model = Sequential()
model.add(LSTM(50, activation='tanh', input_shape=(10, 1)))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_data, y_data, epochs=50, batch_size=16, verbose=1)
```

```python
predictions = model.predict(X_data[:5])
print(predictions)
```

```python
import matplotlib.pyplot as plt
plt.plot(y_data, label='Actual')
plt.plot(model.predict(X_data), label='Predicted')
plt.legend()
plt.show()
```
