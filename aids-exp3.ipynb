```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/aids/StudentsPerformance.csv')
```

```python
print(data.head())
print(data.info())
print(data.columns)
```

```python
data = data.dropna()
data = data.drop(columns=['Unnamed: 0'], errors='ignore')
```

```python
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap using Seaborn')
plt.show()
```

```python
sns.scatterplot(x='math score', y='reading score', data=data)
plt.title('Scatter Plot of Math Score vs Reading Score')
plt.show()
```

```python
sns.lineplot(x='gender', y='parental level of education', data=data)
plt.title('Line Plot using Seaborn')
plt.show()
```

```python
plt.hist(data['parental level of education'], bins=10, color='lightblue', edgecolor='black')
plt.title('Histogram using Matplotlib')
plt.xlabel('Feature Values')
plt.ylabel('Frequency')
plt.show()
```

```python
data['gender'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Pie Chart using Matplotlib')
plt.ylabel('')
plt.show()
```

```python
plt.boxplot(data['math score'])
plt.title('Box Plot using Matplotlib')
plt.show()
```
