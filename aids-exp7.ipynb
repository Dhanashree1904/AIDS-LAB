```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
X = digits.data
y = digits.target
```

```python
df = pd.DataFrame(X)
df['Label'] = y
print(df.head())
```

```python
df.iloc[:, :-1] = df.iloc[:, :-1] / 16.0
```

```python
df.to_csv('digits_preprocessed.csv', index=False)
print("Dataset saved as digits_preprocessed.csv")
```

```python
loaded_df = pd.read_csv('digits_preprocessed.csv')
print(loaded_df.shape)
```
