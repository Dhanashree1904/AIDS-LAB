```python
import numpy as np

A = np.array([0.1, 0.4, 0.7, 0.9])
B = np.array([0.2, 0.5, 0.6, 1.0])

while True:
    print("\n0: Enter new sets")
    print("1: Union")
    print("2: Intersection")
    print("3: Complement of A")
    print("4: Complement of B")
    print("5: Algebraic Sum")
    print("6: Algebraic Product")
    print("7: Bounded Sum")
    print("8: Bounded Difference")
    print("9: Exit")

    choice = int(input("Enter your choice: "))

    if choice == 0:
        A = np.array(list(map(float, input("Enter set A values: ").split())))
        B = np.array(list(map(float, input("Enter set B values: ").split())))

    elif choice == 1:
        print("Union:", np.maximum(A, B))

    elif choice == 2:
        print("Intersection:", np.minimum(A, B))

    elif choice == 3:
        print("Complement of A:", 1 - A)

    elif choice == 4:
        print("Complement of B:", 1 - B)

    elif choice == 5:
        print("Algebraic Sum:", A + B - (A * B))

    elif choice == 6:
        print("Algebraic Product:", A * B)

    elif choice == 7:
        print("Bounded Sum:", np.minimum(1, A + B))

    elif choice == 8:
        print("Bounded Difference:", np.maximum(0, A - B))

    elif choice == 9:
        print("Goodbye! â€“ Dhanashree ðŸ™‚")
        break

    else:
        print("Invalid choice, try again!")
```
